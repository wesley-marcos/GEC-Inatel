### Criação dos dados

```python
from sklearn.datasets import make_blobs

#make_blobs(número de amostras, número de features, número de grupos, razão que define quão grande será o range do grupo, shuffle)
x, y = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=0.5, shuffle=True)
```

### Visualização

```python
import matplotlib.pyplot as plt

#Definindo tamanho da imagem
plt.rcParams['figure.dpi'] = 150
#Definindo os dados do gráfico, como cor, tipo de representação (desenho)
plt.scatter(x[:, 0], x[:, 1], marker="o", edgecolors="black")
plt.show()
```

### Criação do KMeans

```python
from sklearn.cluster import KMeans

#KMeans = modelo de Agrupamento
#KMeans(número de grupos)
modelo = KMeans(n_clusters=3)
#'.fit' fará o aprendizado do método em si
modelo.fit(x)
```

### Predições

```python
#Coloca-se a predição como 'y' porque não se tem o Y original (targets) no Agrupamento
y_km = modelo.predict(x)

#Definindo os conceitos básicos para criação do gráfico
plt.scatter(x[y_km == 0, 0], x[y_km == 0, 1], c="green", marker="s", label="Grupo 1", edgecolors="black")
plt.scatter(x[y_km == 1, 0], x[y_km == 1, 1], c="blue", marker="o", label="Grupo 2", edgecolors="black")
plt.scatter(x[y_km == 2, 0], x[y_km == 2, 1], c="red", marker="v", label="Grupo 3", edgecolors="black")

#Definindo os dados dos centróides
plt.scatter(modelo.cluster_centers_[:, 0], modelo.cluster_centers_[:, 1], c="orange", label="Centros", marker="*")

#Definindo a escala
plt.legend(scatterpoints=1)

plt.show()
```

### Scores

```python
from sklearn.metrics import calinski_harabasz_score, silhouette_score

#Chamando os métodos de Calinski (quanto maior o número, melhor o resultado do método) e Silhouette (porcentagem de precisão do método)
print("CH:", calinski_harabasz_score(x, modelo.labels_))
print("SC:", silhouette_score(x, modelo.labels_))
```
